```python
import sys
sys.path.append('../my/')
from db import pool
import pandas as pd

conn = pool.dedicated_connection()
cursor = conn.cursor()

sql = "select round(p1,4),round(p2,4) from investor.price_change_direction2behavior where samples>=35"
cursor.execute(sql)
df = pd.DataFrame(data=list(cursor.fetchall()), columns=['p1','p2'])

cursor.close()
conn.close()
```

```python
from pyecharts import Scatter
scatter = Scatter("", width=1000, height=618)
scatter.add("", x_axis=df.p1, y_axis=df.p2, xaxis_name="P(|)", yaxis_name="P(|)", 
            xaxis_name_pos="middle", yaxis_name_pos="end", yaxis_name_gap=18)
scatter
```
